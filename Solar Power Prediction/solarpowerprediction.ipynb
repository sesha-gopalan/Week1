```python
import matplotlib.pyplot as plt
```

```python
# Plot distribution of power
# Set overall style and color palette
sns.set_style("whitegrid")
sns.set_palette("viridis")

# Plot distribution of power with enhancements
plt.figure(figsize=(12, 7))
sns.histplot(df['generated_power_kw'], bins=30, kde=True, color="blue")
plt.title('Distribution of Generated Power (kW)', fontsize=16, fontweight='bold')
plt.xlabel('Generated Power (kW)', fontsize=14)
plt.ylabel('Frequency', fontsize=14)
plt.grid(True, linestyle='--', alpha=0.6)

# Add an annotation (optional)
mean_power = df['generated_power_kw'].mean()
plt.axvline(mean_power, color='red', linestyle='--', linewidth=1.5, label=f'Mean: {mean_power:.2f} kW')
plt.legend(fontsize=12)

plt.tight_layout()
plt.show()
```
